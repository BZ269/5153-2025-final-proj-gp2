### Helper functions

```python
def column_names(df):
    first_row = df.loc[0,:]
    columns = []
    for i in first_row:
        columns.append(i[0])
    return columns

def parse_rows(row):
    return {k:v for k,v in row}
```

```python
def convert_datetime(df, gmt=8):
    df['start_time'] = pd.to_datetime(df['start_time'], unit='s') + pd.Timedelta(hours=gmt)
    df['end_time'] = pd.to_datetime(df['end_time'], unit='s') + pd.Timedelta(hours=gmt)
    return df

def filter_datetime(df, column='start_time', filter_dt='2025-04-26 19:30:00'):
    df[column] = pd.to_datetime(df[column])
    df = df[df[column] > filter_dt]
    return df

def latency_ms(df):
    latency = (df['end_time'] - df['start_time']).dt.total_seconds() * 1000
    return latency
```

```python
def get_model_from_extra(nested_dict):
    return nested_dict['metadata']['ls_model_name'] if 'metadata' in nested_dict and 'ls_model_name' in nested_dict['metadata'] else 'NA'

def get_agent_from_tags(tags):
    output = 'NA'
    for tag in tags:
        if tag.startswith('agent'):
            output = tag
    return output

def get_function_from_tags(tags):
    output = 'NA'
    for tag in tags:
        if tag.startswith('function'):
            output = tag
    return output
```

```python
plt.figure(figsize=(4, 6))
plt.title('Length of LangGraph runs by Agent')

df1 = ldf.groupby(['root_run_id', 'agent', 'function'])['total_tokens'].count().reset_index()
df1 = df1[df1['total_tokens'] < 40] # remove outliers
df1 = df1.reset_index(drop=True)

sns.violinplot(x='agent', y='total_tokens', data=df1, palette='Paired', hue='agent', inner=None)

sns.boxplot(
    x='agent', y='total_tokens', data=df1, palette='Paired', width=0.1,
    boxprops={'facecolor':'white', 'edgecolor':'black', "zorder":10}, showfliers=False, 
    whiskerprops={'linewidth':2}, medianprops={'color': 'black', 'linewidth': 2, 'zorder': 20}
)


plt.xlabel('Agent')
plt.ylabel('No. of LangGraph Nodes')
plt.ylim(0)
plt.grid(True, alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(4, 6))
plt.title('Latency of LangGraph runs by Agent')

df2 = ldf.groupby(['root_run_id', 'agent', 'function'])['latency'].sum().reset_index()
df2 = df2[df2['latency'] < 100] # remove outliers
df2 = df2.reset_index(drop=True)

sns.violinplot(x='agent', y='latency', data=df2, palette='Paired', hue='agent', inner=None)

sns.boxplot(
    x='agent', y='latency', data=df2, palette='Paired', width=0.1, 
    boxprops={'facecolor':'white', 'edgecolor':'black', "zorder":10}, showfliers=False, 
    whiskerprops={'linewidth':2}, medianprops={'color': 'black', 'linewidth': 2, 'zorder': 20}
)

plt.xlabel('Agent')
plt.ylabel('Latency (s)')
plt.ylim(0)
plt.grid(True, alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(4, 6))
plt.title('Total Tokens usage by Agent')

df3 = ldf.groupby(['root_run_id', 'agent', 'function'])['total_tokens'].sum().reset_index()
df3 = df3[df3['total_tokens'] < 50000] # remove outliers
df3 = df3.reset_index(drop=True)

sns.violinplot(x='agent', y='total_tokens', data=df3, palette='Paired', hue='agent', inner=None)

sns.boxplot(
    x='agent', y='total_tokens', data=df3, palette='Paired', width=0.1,
    boxprops={'facecolor':'white', 'edgecolor':'black', "zorder":10}, showfliers=False, 
    whiskerprops={'linewidth':2}, medianprops={'color': 'black', 'linewidth': 2, 'zorder': 20}
)

plt.xlabel('Agent')
plt.ylabel('Tokens')
plt.ylim(0)
plt.grid(True, alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.title('Latency vs Total Tokens')

# Prepare your data
df4 = ldf.groupby(['root_run_id', 'model', 'agent', 'function'])[['total_tokens', 'latency']].sum().reset_index()
df4 = df4[df4['latency'] < 100]
df4 = df4[df4['total_tokens'] < 50000]
df4.sort_values(by='model', ascending=False, inplace=True)

# Set color palette
palette = sns.color_palette('deep')  # or pick another like 'tab10'

# Map model to color
models = df4['model'].unique()
model_color_map = {model: palette[i % len(palette)] for i, model in enumerate(models)}

# Scatter plot
sns.scatterplot(data=df4, x='total_tokens', y='latency', hue='model', palette=model_color_map)

# Best fit lines
for model in models:
    sns.regplot(
        data=df4[df4['model'] == model],
        x='total_tokens',
        y='latency',
        scatter=False,
        ci=None,
        color=model_color_map[model],   # matching color
        line_kws={'linestyle': '--', 'linewidth': 2}  # dashed line
    )

plt.xlabel('Prompt Tokens')
plt.ylabel('Latency (s)')
plt.legend()
plt.grid(True, alpha=0.5)
plt.show()
```
